```python
import time
import cv2
import numpy as np
from PIL import ImageGrab
from keras.models import load_model
from keras import backend as K
from key_util import press_key, release_key

def up():
    #release_key('A')
    #release_key('D')
    #release_key('S')
    press_key('W')
    time.sleep(1)
    release_key('W')

def left():
    #release_key('W')
    #release_key('S')
    #release_key('D')
    
    press_key('A')
    time.sleep(1)
    release_key('A')

def right():
    #release_key('W')
    #release_key('A')
    #release_key('S')
    press_key('D')
    time.sleep(1)
    release_key('D')
    
def down():
    #release_key('A')
    #release_key('D')
    #release_key('W')
    press_key('S')
    time.sleep(1)    
    release_key('S')
    
def space(staycount):
    #press_key('Space')
    if staycount%5==0:
        print(staycount)
    elif staycount%5==1:
        up()
    elif staycount%5==2:
        left()
    elif staycount%5==3:
        down()
    elif staycount%5==4:
        right()
    #release_key('A')
    #release_key('D')
    #release_key('S')
    #release_key('W')


def countdown(sec):
    for i in range(sec, 0, -1):
        print(i)
        time.sleep(1)
    print('start!')


def main():
    last_time = time.time()
    frame, accum_time, fps = 0, 0, 0

    model = load_model('model.h5')
    print('Model loading complete!')
    countdown(3)
    count = 0
    staycount=0
    while True:
        # screenshot normalization
        
        screen = np.array(ImageGrab.grab(bbox=(0, 30, 695, 425)))
        gray = cv2.cvtColor(screen, cv2.COLOR_RGB2GRAY)
        gray = cv2.Canny(gray, 50, 150)  # edges detection   

        # get train image
        test_img = gray
        test_img = cv2.resize(test_img, (278, 158))
        
        img_ndarray = np.asarray(test_img, dtype='float64') / 255
        test_data = np.ndarray.flatten(img_ndarray)
        test_data = test_data.astype('float32')

        if K.image_data_format() == 'channels_first':
            test_data = test_data.reshape(1, 1, 158, 278)
        else:
            test_data = test_data.reshape(1,158 , 278, 1)
        
        preds = model.predict(test_data)
        class_ = np.argmax(preds[0])

        if class_ == 0:
            up()
        elif class_ == 1:
            down()
        elif class_ == 2:
            left()
        elif class_ == 3:
            right()
        elif class_ ==4:
            space(staycount)
            staycount+=1
        #count += 1
        #if count == 100:
        #    break

        # show screenshotw
        #cv2.imshow('screen', test_img)w
        #if cv2.waitKey(25) & 0xFF == ord('q'):
        #    cv2.destoryAllWindows()
        #    break


if __name__ == "__main__":
    main()
```
